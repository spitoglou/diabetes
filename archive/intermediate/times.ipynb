```python
from datetime import datetime
from datetime import timedelta
```

```python
time_str = '13/09/2021 12:33:00'
date_format_str = '%d/%m/%Y %H:%M:%S'
start_time = datetime.strptime(time_str, date_format_str)
print('Time Series start time: ', start_time)
```

```python
def rounder(t):
    if t.second >= 30:
        if t.minute < 59:
            return t.replace(second=0, minute=t.minute+1)
        else:
            if t.hour < 23:
                return t.replace(second=0, minute=0, hour=t.hour+1)
            else:
                return t.replace(second=0, minute=0, hour=0, day=t.day+1)
    else:
        return t.replace(second=0)
```

```python
window_start_array = []
window_end_array = []
target_time_array  = []
for index, value in tos.items():
    first = start_time + timedelta(hours=(value))
    first = rounder(first)
    window_start_array.append(first)
for index, value in tol.items():
    last = start_time + timedelta(hours=(value))
    last = rounder(last)
    target = last + timedelta(minutes=30)
    window_end_array.append(last)
    target_time_array.append(target)
```

```python
def get_part_of_day(hour):
    return (
        "morning" if 7 <= hour <= 11
        else
        "afternoon" if 12 <= hour <= 16
        else
        "evening" if 16 <= hour <= 20
        else
        "night" if 21 <= hour <= 23
        else
        "late night"
    )
```

```python
day_part_array = []
for time in window_end_array:
    day_part_array.append(get_part_of_day(time.hour))
day_part_array[:10]
```

```python
import pandas as pd
ws = pd.Series(window_start_array)
we = pd.Series(window_end_array)
tt = pd.Series(target_time_array)
tod = pd.Series(day_part_array)
df = pd.DataFrame({
    'window_start': ws,
    'window_end': we,
    'target_time': tt,
    'time_of_day': tod,
    'value': tdf['label']
})
df.head(130)
```
